```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('twitter_training.csv')
validation_data = pd.read_csv('twitter_validation.csv')
```

```python
train_data.columns = ['ID','Topic','Sentiment','Text']
validation_data.columns = ['ID','Topic','Sentiment','Text']
```

```python
train_data_cleaned = train_data.dropna(subset=['Sentiment','Text'])
validation_data_cleaned = validation_data.dropna(subset=['Sentiment','Text'])
```

```python
plt.figure(figsize=(12, 8))
train_data_cleaned['Sentiment'].value_counts().plot(kind='bar', color=['black','grey','brown','blue'])
plt.title('Sentiment Distribution in Training Data')
plt.xlabel('Sentiment')
plt.ylabel('Number of Entries')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
validation_data_cleaned['Sentiment'].value_counts().plot(kind='bar', color=['black','grey','brown','blue'])
plt.title('Sentiment Distribution in Validation Data')
plt.xlabel('Sentiment')
plt.ylabel('Number of Entries')
plt.xticks(rotation=0)
plt.show()
```
